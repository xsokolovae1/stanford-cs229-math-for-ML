```python
import numpy as np
import matplotlib.pyplot as plt
```

Implement BackPropagation for AND problem

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]).T
d = np.array ([0, 0, 0, 1])
```

```python
def network_parameters():

  input_size = 2
  hidden_size = 2
  output_size = 1

  lr = 0.01
  epochs = 10000


  w1 = np.random.rand(hidden_size, input_size) * 2 - 1
  b1 = np.random.rand(hidden_size, 1) * 2 - 1
  w2 = np.random.rand(output_size, hidden_size) * 2 - 1
  b2 = np.random.rand(output_size, 1) * 2 - 1



  return w1, b1, w2, b2, lr, epochs
```

```python
w1, b1, w2, b2, lr, epochs = network_parameters()
error_list = []
```

```python
for epoch in range (epochs):
  #forward pass
  z1 = np.dot(w1, X) + b1
  #sigmoid
  a1 = 1/ (1 + np.exp(-z1))

  #do the same the second time

  z2 = np.dot(w2, X) + b2
  a2 = 1/(1+np.exp(-z2))


  #error
  e = d - a2
  #derivatedie
  dz2 = e * (a2 * (1 - a2))


  #backprrop
  da1 = np.dot(w2.T, dz2)
  dz1 = da1 * (a1 * (1 - a1))


  #updates of w and b

  w2 += lr * np.dot(dz2, a1.T)
  b2 += lr * np.sum(dz2, axis=1, keepdims=True)


  w1 += lr * np.dot(dz1, X.T)
  b1 += lr * np.sum(dz1, axis=1, keepdims=True)

  if (epoch+1)%1000 == 0:
      print("Epoch: %d, Average error: %0.05f"%(epoch, np.average(abs(e))))
      error_list.append(np.average(abs(e)))
```

```python
z1 = np.dot(w1, X) + b1
a1 = 1 / (1 + np.exp(-z1))
z2 = np.dot(w2, a1) + b2
a2 = 1 / (1 + np.exp(-z2))


print('Final output after training:', a2)
print('Ground truth', d)
print('Error after training:', e)
print('Average error: %0.05f'%np.average(abs(e)))
```
